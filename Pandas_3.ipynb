```python
df_agg=df.groupby(["director_name"])[["title","year"]].aggregate({"title":"count","year":["min","max"]})
df_agg
```

```python
df_agg["year"]
```

```python
df_agg["year"]["min"]
```

```python
df_agg["year"]["max"]
```

```python
df_agg.columns
```

```python
df_agg.columns=["_".join(col) for col in df_agg.columns]
df_agg
    
```

```python
df_agg.reset_index(inplace=True)
df_agg
```

```python
df_agg["years_active"]=df_agg["year_max"]-df_agg["year_min"]
df_agg
```

```python
df_agg["movies_per_year"]=df_agg["title_count"]/df_agg["years_active"]
```

```python
df_agg
```

```python
df_agg.sort_values(["movies_per_year"],ascending=False,inplace=True)
```

```python
def sub_avg(x):
    x["budget"] =x["budget"]-x["budget"].mean()
df.groupby(["director_name"]).transform(sub_avg)
```

```python
def sub_avg(x):
    x=x-x.mean()
    return x
df.groupby(["director_name"])["budget"].transform(sub_avg)
```
