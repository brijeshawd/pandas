```python
import seaborn as sns
def function(df,r):
  samples = []
  for i in range(r):
    temp_df = df.sample(100,replace=True)
    samples.append(temp_df.mean())
  samples = np.sort(samples)
  return samples

sample_metrics = function(df,1000)

sns.distplot(sample_metrics)
```
