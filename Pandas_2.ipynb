```python
users=pd.DataFrame({"userid":[1,2,3],"name":["Sandeep","Rishabh","Pradeep"]})
users
```

```python
msgs=pd.DataFrame({"userid":[1,1,2],"msg":["Hmmm","Acha","thik_hai"]})
msgs
```

```python
users.merge(msgs,on="userid")
```

```python
users.rename(columns={"userid":"id"},inplace=True)
users
```
